NUMPY PROBLEMS


```python
'''1.convert the mpg column into a numpy array and calculate:
    the mean, median and standard deviation of mpg.
    the number of cars with mpg greater than 25 '''

import numpy as np
import pandas as pd

d=pd.read_csv("auto.csv")

mpg = d['mpg'].to_numpy()

mean = np.mean(mpg)
median = np.median(mpg)
std = np.std(mpg)

count= np.sum(mpg > 25)

print(f"mean={mean},\nmedian={median},\nstd={std},\ncount={count}")
```

```python
'''2.Using NumPy, filter all cars with more than 6 cylinders.
Return the corresponding car_name as a list '''

cylinders = d['cylinders'].to_numpy()
carnames = d['car name'].to_numpy()
cars=carnames[cylinders > 6]

print(f"car list={cars.tolist()}")

```

```python
'''3.Increase all horsepower values by 10% and store the updated values in a new
NumPy array. Handle missing data (if any) by replacing it with the mean of the
column before applying the increase.'''

hp = d['horsepower'].replace('?', np.nan).astype(float).to_numpy()

mean= np.nanmean(hp)
hp = np.where(np.isnan(hp), mean, hp)

hpinc = hp * 1.10

print(f"hp_updated={hpinc}")
```

```python
'''4.Find the average displacement of cars with an origin of 2 (Europe) using NumPy
indexing.'''

dis = d['displacement'].to_numpy()
origin = d['origin'].to_numpy()

avg = np.mean(dis[origin == 2])

print(f"average={avg}")
```

```python
'''5.Calculate the mean mpg for cars grouped by the number of cylinders using NumPy
techniques. '''

cyl = d['cylinders'].to_numpy()
mpg = d['mpg'].to_numpy()

u = np.unique(cyl)

for c in u:
    mean = np.mean(mpg[cyl == c])
    print(f"Cylinders {c}: Mean MPG = {mean}")
```

PANDAS PROBLEM

```python
'''6.Load the dataset into a Pandas DataFrame. Display:
○ The first 10 rows
○ The total number of rows and columns
○ Summary statistics for numerical columns'''

print(f"10 rows\n{d.head(10)}")

print(f"\nnumber of rows and columns\n{d.shape}")

print(f"statistics=\n{d.describe()}")
```

```python
'''7.Find all cars manufactured in 1975 with a weight less than 3000. Return the
DataFrame with selected columns: car_name, weight, and mpg.'''

filter = d[(d['model year'] == 75) & (d['weight'] < 3000)]

result = filter[['car name', 'weight', 'mpg']]

print(f"Details =\n{result}")
```

```python
''' 8.Identify if there are any missing values in the dataset. Replace missing values in the
horsepower column with the column's median. '''

print(d.isnull().sum())

d['horsepower'] = d['horsepower'].replace('?', np.nan).astype(float)

median_hp = d['horsepower'].median()

d['horsepower'].fillna(median_hp, inplace=True)
```

```python
'''9.Add a new column power_to_weight_ratio, calculated as horsepower / weight. '''
d['power_to_weight_ratio'] = d['horsepower'] / d['weight']
```

```python
'''10.Group the cars by origin and calculate the mean mpg for each group. '''

g = d.groupby('origin')['mpg'].mean()

print(f"grouped data\n{g}")
```

```python
'''11.Generate a summary DataFrame with:
○ Average mpg, weight, and horsepower for each model_year. '''

summary = d.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()

print(summary)
```

```python
'''12.Save a subset of the data containing only mpg, cylinders, horsepower, and weight
for cars with mpg > 30 into a CSV file named high_mpg_cars.csv. '''

high = d[d['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]

high.to_csv("high_mpg_cars.csv", index=False)
```
